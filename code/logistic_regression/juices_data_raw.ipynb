# Juices dataset (raw data) 

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import arviz as az

az.style.use('arviz-darkgrid')
```

```python
# read .csv file into Dataframe
data = pd.read_csv('../../data/juices/FTIR_juices.csv')
data.head(3)
```

```python
data.tail(3)
```

```python
data.describe()
```

```python
# get header values as x-axis values for plotting
cols = data.columns.to_list()
x_val = cols[:-1]
```

```python
# shuffle data rows (for plotting a sample fraction)
ndata = data.sample(frac=1)
```

```python
# plot all data in 1 figure
fig = plt.figure(figsize=(15,8))
ax = plt.axes()
ax.xaxis.set_major_locator(plt.MaxNLocator(10))
ax.set(title='Raw FTIR spectrum of juices samples', 
       xlabel='Wavenumber ($cm^{-1}$)', ylabel='Intensity')

plt.rc('axes', labelsize=16)
plt.rc('axes', titlesize=18)

for i in range(100):
    y = ndata.values[i]
    if y[-1] == 'Strawberry':
        col = 'blue'
    else:
        col = 'black'
    y_val = y[:-1]
    plt.plot(x_val, y_val, color=col);
```
